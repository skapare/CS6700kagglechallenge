### Calculating noise in data

```python
start_read = time.time()
df = pd.read_csv("./train_sample.csv")
print(time.time() - start_read)
df
```

```python
import numpy as np
import math

noise_calculator = df.groupby(['ip', 'os', 'device', 'app', 'channel']).agg({'click_time':['count'], 'is_attributed': ['sum']})

def collect_noise(grouped_data):
    ret = []
    ambiguous_points = 0
    for index, row in grouped_data.iterrows():
        average_label = row['is_attributed']['sum'] * 1.0 / row['click_time']['count']
        if average_label != 0 and average_label != 1:
            ambiguous_points += row['click_time']['count']
        for i in range(row['is_attributed']['sum']):
            ret.append(1 - average_label)
        for i in range(row['click_time']['count'] - row['is_attributed']['sum']):
            ret.append(0 - average_label)
        #ret.append((row['is_attributed']['sum']*(average_label - 1) + (row['click_time']['count'] - row['is_attributed']['sum']) * abs(average_label))/row['click_time']['count'])
    return ambiguous_points, ret

num_ambiguous_points, arr = collect_noise(noise_calculator)
arr = np.array(arr)
print("%% ambiguous points: " + str(num_ambiguous_points*1.0/len(df)))
print("average noise: " + str(np.mean(arr, axis=0)))
print("stdev noise: " + str(np.std(arr, axis=0)))
```
